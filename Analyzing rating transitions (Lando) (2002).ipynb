## Аннотация
Мы рассматриваем оценку переходов кредитных рейтингов на основе непрерывных наблюдений. 

На простых примерах и с использованием большого набора данных Standard and Poor's мы иллюстрируем разницу между оценками, основанными на методах дискретной когорты, и оценками, основанными на непрерывных наблюдениях. 

Мы применяем методы полупараметрической регрессии для проверки двух типов немарковских эффектов при переходе от одного рейтинга к другому: зависимость от продолжительности и зависимость от предыдущего рейтинга.

Мы находим значительные немарковские эффекты, особенно для переходов на понижение рейтинга.

# 1. Введение

Матрицы перехода (Матрицы миграции) занимают центральное место в современном управлении кредитными рисками. 

Отчеты о миграции рейтингов, публикуемые агентствами Standard and Poor's и Moody's, изучаются менеджерами по управлению кредитными рисками повсеместно, а некоторые из наиболее известных инструментов управления рисками, такие как Credit Metrics от J.P. Morgan и Credit Portfolio View от McKinsey, построены на оценках вероятности миграции рейтингов. 

По сути, оценки, публикуемые этими агентствами и в научной литературе, используют дискретное время и опираются на "когортный" метод, который оценивает коэффициенты перехода следующим образом: 
+ Учитывая, что в начале года в данной рейтинговой категории $i$ находится $N_i$ компаний-заемщиков и что из этой совокупности $N_{ij}$ перешли в категорию $j$, годовой коэффициент перехода (one year transition rate) оценивается как 
$$
\hat{p}_{ij} = \frac {N_{ii}} {N_i},       j\neq i
$$


Важным следствием этого является то, что если переход из $i$ в $j$ не происходит в данном периоде, то оценка соответствующего коэффициента равна 0.

Рейтинговые агентства, разумеется, имеют доступ к непрерывным данным о переходе рейтингов и знают точные даты в течение года, когда компания меняет свой рейтинг или понижается в нем. 

Аналогично, банк, использующий внутреннюю рейтинговую систему, будет иметь доступ к полной истории рейтинговых переходов. 

Далее мы утверждаем, что для получения эффективных оценок коэффициентов перехода (transition rates) крайне важно основывать оценку коэффициентов перехода на этих непрерывно наблюдаемых историях. 

Этот момент особенно важен при оценке "редких событий", таких как переход от, скажем, ААА в рейтинге Standard and Poor's к дефолту. 

Если говорить очень кратко, то оценка максимального правдоподобия, которую можно получить для годовой вероятности перехода от ААА к дефолту, будет (и должна быть) ненулевой, даже если в период наблюдения не было ни прямых, ни косвенных дефолтов (т. е. дефолтов через последовательность понижений рейтингов).

Кратко говоря, если в течение одного года нет переходов от ААА к дефолту, но есть переходы от ААА к АА и от АА к дефолту (но другими компаниями-заемщиками), то оценка для переходов от ААА к дефолту должна быть ненулевой, поскольку очевидно, что существует вероятность дефолта в течение года после последовательного понижения рейтинга, даже если это не произошло ни с одной фирмой в выборке. 

> Оценка в непрерывном времени позволяет уловить это, в то время как метод дискретного времени этого не делает.
